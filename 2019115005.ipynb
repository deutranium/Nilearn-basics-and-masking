```python
from nilearn import plotting
```

Plotting the anatomical image

```python
plotting.plot_anat(anatomical)
plotting.show()
```

Tweaking `display_mode` for selective results

```python
print("display_mode = x")
plotting.plot_anat(anatomical, display_mode="x")
plotting.show()

print("display_mode = y")
plotting.plot_anat(anatomical, display_mode="y")
plotting.show()

print("display_mode = z")
plotting.plot_anat(anatomical, display_mode="z")
plotting.show()

print("display_mode = yx")
plotting.plot_anat(anatomical, display_mode="yx")
plotting.show()

print("display_mode = xz")
plotting.plot_anat(anatomical, display_mode="xz")
plotting.show()

print("display_mode = yz")
plotting.plot_anat(anatomical, display_mode="yz")
plotting.show()

print("display_mode = ortho")
plotting.plot_anat(anatomical, display_mode="ortho")
plotting.show()

print("display_mode = tiled")
plotting.plot_anat(anatomical, display_mode="tiled")
plotting.show()

print("display_mode = mosaic")
plotting.plot_anat(anatomical, display_mode="mosaic")
plotting.show()
```
